````python
# Dependencies
import numpy as np
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize
import torch
from torch.autograd import Variable
import torch.nn.functional as F
nltk.download('punkt')
````

### Prepare the Data

1.1 Spend some time on preparing the dataset. It may be helpful to lower-case the data and to filter for German titles. The format of the CSV-file should be:

```
Anmeldedatum;Abgabedatum;JahrAkademisch;Art;Grad;Sprache;Titel;Abstract
```

1.2 Create the vocabulary from the prepared dataset. You'll need it for the initialization of the matrices and to map tokens to indices.

1.3 Generate the training pairs with center word and context word. Which window size do you choose?

````python
def load_theses_dataset(filepath):
    """Loads all theses instances and returns them as a dataframe."""
    ### YOUR CODE HERE
    
    df = pd.read_csv(filepath, sep=';')
    return df
    ### END YOUR CODE
````

````python
def preprocess(dataframe):
    """Preprocesses and tokenizes the given theses titles for further use."""
    ### YOUR CODE HERE
    german = dataframe[dataframe["Sprache"]=="DE"]
    german.loc[:, 'Titel'] = german['Titel'].str.lower()
    tokens = [word_tokenize(title) for title in german["Titel"]]
   # tokens = [[word.replace("<br>", "") for word in title] for title in tokens]
    return tokens

def build_vocabulary(tokens):
    flattened_list = [word for title in tokens for word in title]
    vocabulary = set(flattened_list)

    word2idx = {word: idx for idx, word in enumerate(vocabulary)}
    idx2word = {idx: word for word, idx in word2idx.items()}
    return vocabulary, word2idx, idx2word
    ### END YOUR CODE
````

````python
def create_training_pairs(data, word2idx, window_size):
    """Creates training pairs based on skip-grams for further use."""
    ### YOUR CODE HERE
    idx_pairs = []
    for title in data:
        indices=[word2idx[word] for word in title]
        for center_word_pos in range(len(indices)):
            for w in range(-window_size, window_size + 1):
                context_word_pos = center_word_pos + w
                if context_word_pos < 0 or context_word_pos >= len(indices) or center_word_pos == context_word_pos:
                    continue
                context_word_idx = indices[context_word_pos]
                idx_pairs.append((indices[center_word_pos], context_word_idx))

    idx_pairs = np.array(idx_pairs)

    return idx_pairs
    
    ### END YOUR CODE
````

````python
dataframe = load_theses_dataset("./data/theses2022.csv")
preprocessed_data = preprocess(dataframe)
vocabulary, word2idx, idx2word = build_vocabulary(preprocessed_data)

print(list(vocabulary)[:5])
print(len(vocabulary))

print(idx2word[2])
print(word2idx["it-management-tools"])

training_pairs = create_training_pairs(preprocessed_data, word2idx, 2)

for pair in training_pairs[:5]:
    print(f"{idx2word[pair[0]]} {idx2word[pair[1]]}")
````

### Train and Analyze

2.1 Implement and train the word2vec model with your training data.

2.2 Implement a method to find the top-k similar words for a given word (token).

2.3 Analyze: What are the most similar words to "Konzeption", "Cloud" and "virtuelle"?

````python
### TODO: 2.1 Implement and train the word2vec model.

### YOUR CODE HERE

def get_input_layer(word_idx):
    x = torch.zeros(len(vocabulary)).float()
    x[word_idx] = 1.0
    return x

def train_word2vec():
    embedding_dims = 5
    vocabulary_size = len(vocabulary)
    W1 = Variable(torch.randn(embedding_dims, vocabulary_size).float(), requires_grad=True)
    W2 = Variable(torch.randn(vocabulary_size, embedding_dims).float(), requires_grad=True)
    num_epochs = 100
    learning_rate = 0.001
    
    for epo in range(num_epochs):
        loss_val = 0
        for data, target in training_pairs:
            x = Variable(get_input_layer(data)).float()
            y_true = Variable(torch.from_numpy(np.array([target])).long())
    
            z1 = torch.matmul(W1, x)
            z2 = torch.matmul(W2, z1)
        
            log_softmax = F.log_softmax(z2, dim=0)
    
            loss = F.nll_loss(log_softmax.view(1,-1), y_true)
            loss_val += loss.item()
            loss.backward()
            W1.data -= learning_rate * W1.grad.data
            W2.data -= learning_rate * W2.grad.data
    
            W1.grad.data.zero_()
            W2.grad.data.zero_()
        if epo % 10 == 0:    
            print(f'Loss at epo {epo}: {loss_val/len(training_pairs)}')
    
            # Extracting word vectors
    vectors_v_W1 = W1.detach().numpy().T
    vectors_u_W2 = W2.detach().numpy()
    vectors_average = 0.5 * (vectors_v_W1 + vectors_u_W2)

    return vectors_v_W1, vectors_u_W2, vectors_average

vectors_v_W1, vectors_u_W2, vectors_average = train_word2vec()
torch.save({'W1': vectors_v_W1, 'W2': vectors_u_W2}, 'word2vec_weights.pth')

### END YOUR CODE
````

````python
### TODO: 2.2 Implement a method to find the top-k similar words.

### YOUR CODE HERE

from sklearn.metrics.pairwise import cosine_similarity

def find_top_similar_words(target_word, word_vectors, vocabulary, word2idx, idx2word, top_k=5):
    if target_word not in vocabulary:
        print(f"Word '{target_word}' not found in the vocabulary.")
        return []

    target_idx = word2idx[target_word]
    target_vector = word_vectors[target_idx].reshape(1, -1)
    
    similarities = cosine_similarity(target_vector, word_vectors).flatten()
    
    # Sort similarities and get top-k indices
    top_indices = similarities.argsort()[-top_k:][::-1]
    
    # Return top-k similar words
    similar_words = [(idx2word[i], similarities[i]) for i in top_indices]
    return similar_words


### END YOUR CODE
````

````python
### TODO: 2.3 Find the most similar words for "Konzeption", "Cloud" and "virtuelle".

### YOUR CODE HERE

def load_word2vec_weights(file_path):
    checkpoint = torch.load(file_path)
    
    vectors_v_W1 = checkpoint['W1']
    vectors_u_W2 = checkpoint['W2']
    
    return vectors_v_W1, vectors_u_W2

vectors_v_W1, vectors_u_W2 = load_word2vec_weights('word2vec_weights.pth')

konzeption = find_top_similar_words("konzeption", vectors_u_W2, vocabulary, word2idx, idx2word)
cloud = find_top_similar_words("cloud", vectors_u_W2, vocabulary, word2idx, idx2word)
virtuelle = find_top_similar_words("virtuelle", vectors_u_W2, vocabulary, word2idx, idx2word)

print(konzeption)
print(cloud)
print(virtuelle)

### END YOUR CODE
````

### Play with the Embeddings

3.1 Use the computed embeddings: Can you identify the most similar theses for some examples?

3.2 Visualize the embeddings for a subset of theses using [TSNE](https://scikit-learn.org/stable/modules/generated/sklearn.manifold.TSNE.html). You can use [Scikit-Learn](https://scikit-learn.org/stable/) and [Matplotlib](https://matplotlib.org) or [Seaborn](https://seaborn.pydata.org).

````python
### TODO: 3.1 Compute the embeddings for the theses and transform with TSNE.

### YOUR CODE HERE

import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

def average_word_embeddings(title_words, word_embeddings, word2idx):
    """
    Calculate the average embedding of a list of words.
    """
    embeddings = [word_embeddings[word2idx[word]] for word in title_words if word in word2idx]
    if len(embeddings) == 0:
        return None
    return np.mean(embeddings, axis=0)

def find_most_similar_theses(target_thesis_title, thesis_titles, thesis_word_embeddings, word2idx, idx2word, top_k=5):
    if target_thesis_title not in thesis_titles:
        print(f"Thesis '{target_thesis_title}' not found.")
        return []

    target_word_indices = [word2idx[word] for word in target_thesis_title if word in word2idx]
    if not target_word_indices:
        print(f"No embeddings found for the words in '{target_thesis_title}'.")
        return []

    target_word_embeddings = thesis_word_embeddings[target_word_indices]
    target_embedding = np.mean(target_word_embeddings, axis=0)

    avg_embeddings = []
    for title in thesis_titles:
        title_word_indices = [word2idx[word] for word in title if word in word2idx]
        if title_word_indices:
            title_word_embeddings = thesis_word_embeddings[title_word_indices]
            avg_embedding = np.mean(title_word_embeddings, axis=0)
            avg_embeddings.append(avg_embedding)
        else:
            avg_embeddings.append(None)

    similarities = cosine_similarity([target_embedding], avg_embeddings).flatten()
    top_indices = similarities.argsort()[-top_k-1:-1][::-1]  # Get indices of top-k similar theses

    similar_theses = [(thesis_titles[i], similarities[i]) for i in top_indices if avg_embeddings[i] is not None]
    return similar_theses

target_thesis_title = word_tokenize("Design und Implementierung eines objektorientierten Kommunikations-Framework".lower())
top_similar_theses = find_most_similar_theses(target_thesis_title, preprocessed_data, vectors_u_W2, word2idx, idx2word, top_k=5)
print(f"Most similar theses to '{' '.join(target_thesis_title)}':")
for thesis, similarity in top_similar_theses:
    print(f"{' '.join(thesis)}: {similarity}")
### END YOUR CODE
````

````python
### TODO: 3.2 Visualize the samples in the 2D space.

### YOUR CODE HERE

import numpy as np
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt

def visualize_embeddings(thesis_titles, thesis_word_embeddings, word2idx, idx2word, subset_size=100):
    subset_titles = thesis_titles[:subset_size]
    subset_embeddings = thesis_word_embeddings[:subset_size]

    avg_embeddings = []
    for title in subset_titles:
        title_word_indices = [word2idx[word] for word in title if word in word2idx]
        if title_word_indices:
            title_word_embeddings = thesis_word_embeddings[title_word_indices]
            avg_embedding = np.mean(title_word_embeddings, axis=0)
            avg_embeddings.append(avg_embedding)
        else:
            avg_embeddings.append(None)

    avg_embeddings = [emb for emb in avg_embeddings if emb is not None]

    tsne = TSNE(n_components=2, perplexity=10, random_state=42)
    embeddings_2d = tsne.fit_transform(np.array(avg_embeddings))  

    # Plot the embeddings
    plt.figure(figsize=(10, 8))
    for i, (x, y) in enumerate(embeddings_2d):
        plt.scatter(x, y, marker='o', color='b')
        plt.text(x, y, ' '.join(subset_titles[i]), fontsize=9)

    plt.title('t-SNE Visualization of Thesis Title Embeddings')
    plt.xlabel('t-SNE Dimension 1')
    plt.ylabel('t-SNE Dimension 2')
    plt.show()

visualize_embeddings(preprocessed_data, vectors_u_W2, word2idx, idx2word, subset_size=20)

### END YOUR CODE
````
